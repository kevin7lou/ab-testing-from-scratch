```python
from math import *
from random import randint
import matplotlib.pyplot as plt
%matplotlib inline

def iter(li):
    n = len(li)
    for i in xrange(n):
        if li[i] == 0: continue
        j = randint(0, n - 1)
        # the i-th person gives $1 to the j-th person
        li[i] -= 1
        li[j] += 1

wealth_history = []
pareto_history = []
def run(li, num_iter):
    for i in xrange(num_iter):
        iter(li)
        wealth_history.append(li[0])
        pareto = float(sum(sorted(li)[int(0.8*num_people):num_people]))/(num_people*initial_money)
        pareto_history.append(pareto)
        if i % 100 == 0: print(i)

num_people = 1000
initial_money = 100
num_iter = 1*1000

li = [initial_money] * num_people
run(li, num_iter)

plt.bar(range(num_people), sorted(li))
plt.xlabel('Rank')
plt.ylabel('Wealth')
plt.title(r'Sorted wealth')
plt.grid(True)
plt.show()
n, bins, patches = plt.hist(li, 50)
plt.xlabel('Wealth')
plt.ylabel('# People')
plt.title(r'Distribution of people according to wealth')
plt.grid(True)
plt.show()
n, bins, patches = plt.hist(li, 50, weights=li)
plt.xlabel('Wealth')
plt.ylabel('Total wealth')
plt.title(r'Distribution of wealth according to wealth')
plt.grid(True)
plt.show()
plt.plot(wealth_history)
plt.xlabel('Time')
plt.ylabel('Wealth')
plt.title(r'The cycle of wealth')
plt.grid(True)
plt.show()
plt.plot(pareto_history)
plt.xlabel('Time')
plt.ylabel('% wealth held by the top 20%')
plt.title(r'Wealth held by the top 20%')
plt.grid(True)
plt.show()
```
