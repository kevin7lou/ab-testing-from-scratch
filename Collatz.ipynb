```python
from math import log2
import matplotlib.pyplot as plt
%matplotlib inline

def collatz(n):
    if n % 2 == 0: return int(n/2)
    else:          return 3*n+1

def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        n = collatz(n)
        sequence += [n]
    return sequence

def binary(n): return "{0:b}".format(n)

def collatz_sequence_bits(n):
    ["{0:b}".format(i) for i in collatz_sequence(n)]

def max_binary1s(n):
    return collatz_sequence_bits(n).count('1')
    
def max_power2(sequence):
    for i in sequence:
        if i != 0 and ((i & (i - 1)) == 0): return i        
```
